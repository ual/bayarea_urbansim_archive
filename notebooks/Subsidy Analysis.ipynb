```python
import pandas as pd
```

```python
df = pd.read_csv("run41_subsidy_summary.csv")
df2 = pd.read_csv("run41_parcel_output.csv")
df3 = pd.read_csv("run42_parcel_output.csv")
```

```python
df = df.query("amount > 0")
grps = df.groupby("year")
df.head(3)
```

```python
out = pd.DataFrame({
    "subdidy_in_millions": (grps.amount.sum()/-1000000).astype('int'),
    "subsidized_units": grps.residential_units.sum(),
    "units_per_building": grps.residential_units.mean().astype('int'),
    "subdidy_per_unit": (grps.amount.sum()/100/grps.residential_units.sum()*-1).astype('int')*100,
    "total_units": df2.groupby('year_built').residential_units.sum(),
    "total_units_scenario2": df3.groupby('year_built').residential_units.sum()
})
out
```

```python
print out.sum()
print out.mean()
```
